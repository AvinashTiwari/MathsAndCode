```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shift = 5
stretch = 2
n =1000

#pnts = np.random.randn(n)*stretch +shift
pnts = np.random.randn(n)*stretch
print('Mean is %g, std %g' %(np.mean(pnts), np.std(pnts)))
```
