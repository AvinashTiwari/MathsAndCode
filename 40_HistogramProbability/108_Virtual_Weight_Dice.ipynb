```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
outcomes = np.zeros(6)
pureprobs = np.array([2,1,1,1,1,1])
purepobs = pureprobs / np.sum(pureprobs)
purepobs
```

```python
for i in range(n):
    dieroll = np.random.randint(0,7)
    if dieroll == 6:
        dieroll = 0
        
    outcomes[dieroll] +=  1

plt.bar(range(1,7),outcomes/n, color='k', label='Emprical (proportion)')
plt.plot(range(1,7),pureprobs, 'ms-', markersize=10 ,label='Emprical (prob)')
plt.xlabel('Die Face')
plt.ylabel('Probabiblty or Proportion')
plt.legend()
for i in range(1,7):
    plt.text(i,outcomes[i-1]/n-0.25, '%s%%' %int((100*outcomes[i-1])/n) , color="w", fontsize=12,horizontalalignment='center')
    
```
