```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,2,3,4,5,6,6,6,6,7,8,9,8,6,3]
plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of apperance by count')
plt.show()
```

```python
histdata = np.histogram(data, bins=np.arange(1,11))
print(histdata)
plt.bar(histdata[1][:-1], histdata[0])
plt.xlabel('Value')
plt.ylabel('Number of apperance by count')
plt.show()
```

```python
y,x = np.histogram(data, bins=np.arange(1,11))
yp = y/np.sum(y)
plt.bar(x[:-1], yp)
plt.xlabel('Value')
plt.ylabel('Propotion ')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
y,x = np.histogram(data, bins=np.arange(1,11))
yp = y/np.max(y)
plt.bar(x[:-1], yp)
plt.xlabel('Value')
plt.ylabel('Propotion norm')
plt.xticks(np.arange(1,10))
plt.show()
```
