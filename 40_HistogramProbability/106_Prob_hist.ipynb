```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,2,3,4,5,6,6,6,6,7,8,9,8,6,3]
plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of apperance by count')
plt.show()
```

```python
histdata = np.histogram(data, bins=np.arange(1,11))
print(histdata)
plt.bar(histdata[1][:-1], histdata[0])
plt.xlabel('Value')
plt.ylabel('Number of apperance by count')
plt.show()
```

```python
y,x = np.histogram(data, bins=np.arange(1,11))
yp = y/np.sum(y)
plt.bar(x[:-1], yp)
plt.xlabel('Value')
plt.ylabel('Propotion ')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
y,x = np.histogram(data, bins=np.arange(1,11))
yp = y/np.max(y)
plt.bar(x[:-1], yp)
plt.xlabel('Value')
plt.ylabel('Propotion norm')
plt.xticks(np.arange(1,10))
plt.show()
```

```python
x =  np.linspace(-3,3, 1000)
fig, ax = plt.subplots(2,3)
fig.set_size_inches(10,6)
for i in range(3):
    y = x**(i+1)
    h = np.histogram(y,100)
    yp = h[0]/np.sum(h[0])
    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^{%g}$'%(i+1))
    ax[1, i].plot(h[1][:-1], yp)
```
