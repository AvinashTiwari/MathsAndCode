```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)
nums = nums +3
plt.plot(nums , 's')
plt.show()
```

```python
plt.hist(nums )
plt.show()
```

```python
nums = np.random.rand(1000)
nums = nums +3
minval = 2
maxval = 17
nums = nums*(maxval-minval) + minval
plt.plot(nums , 's')
plt.show()
plt.hist(nums )
plt.show()
```

```python
nums = np.random.randn(1000)

plt.plot(nums , 's', alpha=0.5)
plt.show()
plt.hist(nums )
plt.show()
```

```python
desired_mean = 15
desired_std = 4.3
nums = np.random.randn(1000) * desired_std + desired_mean
plt.subplot(121)
plt.plot(nums,'s')
plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean 0f %s and standard deviation %s ' %(np.mean(nums), np.std(nums)))
```
