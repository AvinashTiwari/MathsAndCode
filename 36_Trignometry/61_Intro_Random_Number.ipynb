```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)
nums = nums +3
plt.plot(nums , 's')
plt.show()
```

```python
plt.hist(nums )
plt.show()
```

```python
nums = np.random.rand(1000)
nums = nums +3
minval = 2
maxval = 17
nums = nums*(maxval-minval) + minval
plt.plot(nums , 's')
plt.show()
plt.hist(nums )
plt.show()
```

```python
nums = np.random.randn(1000)

plt.plot(nums , 's', alpha=0.5)
plt.show()
plt.hist(nums )
plt.show()
```
