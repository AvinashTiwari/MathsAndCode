```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
k = np.pi / 4
m = 2.3

eul = m*np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))

print(eul)
print(cis)
```

```python
mag =np.abs(eul)
ang = np.angle(eul)
plt.polar([0,ang], [0,mag], 'r')
plt.polar(k,m, 'bo')
```

```python
def eulerFromCosSine():
    re = eval (input("cosinse part: "))
    im = eval (input(" sine part :"))
    m = np.sqrt(re**2 + im**2)
    k = np.arctan2(im,re)
    plt.polar([0,k], [0,m])
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$=%g' %(m,k))
    plt.show()
```

```python
nvects = 200
ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi
for i in range(0,nvects):
    r = np.random.rand()
    
    if r < .4:
        clr = [1,.2,.7]
    elif  r > .4 and r < .8:
        clr = [.7,.2,1]
    else:
        clr = [0,1,0]
        
    
    plt.polar([0,ks[i]], [0, ms[i]], color = clr)
    
plt.axis('off')
```

```python
t = np.linspace(0,20*np.pi, 1000)
r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

x = np.append(x, x[0]) 
y = np.append(y, y[0])
plt.plot(x,y, 'k')
plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.title('r$_1$=%s, r$_2$=%s' %(np.round(r1,2), np.round(r2,2)))
plt.axis('off')
plt.show()
```
