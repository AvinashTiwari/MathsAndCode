```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
k = np.pi / 4
m = 2.3

eul = m*np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))

print(eul)
print(cis)
```

```python
mag =np.abs(eul)
ang = np.angle(eul)
plt.polar([0,ang], [0,mag], 'r')
plt.polar(k,m, 'bo')
```
