```python

import sympy as sym
import sympy.plotting as symplot
```

```python
from sympy.abc import x
y = x**2
```

```python

p = symplot.plot(y, show=False)
p.ylim =[0,50]
p.xlim =[0,50]
p[0].line_color= 'm'
p.title = "This is graph sample"
p.show()
```

```python
x,a = sym.symbols('x,a')
```

```python
expr = a/x
n = 1
p = symplot.plot(expr.subs(a,n), (x,-5,5), show=False)
p2 = symplot.plot(expr.subs(a,n+1), (x,-5,5), show=False)
p[0].label = '$y=%s$'%sym.latex(expr.subs(a,n))
p[0].line_color ='r'
p.extend(p2)
p.ylim =[-5,5]
p.legend = True
p.show()
```
