```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
y = np.array([[1,1],[2,3],[3,1]])
y1 = np.array([[2,2],[2.5,4],[3.5,1]])
y
```

```python
p = Polygon(y, facecolor='m', alpha=.3)
p1 = Polygon(y1, facecolor='b', alpha=.9, edgecolor='k')
fig,ax = plt.subplots()




ax.add_patch(p)
ax.add_patch(p1)

ax.set_xlim([0,4])
ax.set_ylim([0,4])
plt.show()
```

```python
x = np.linspace(-2,2,101)
f = -x**2
pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g', alpha=0.2)
y = np.array([[-.5,-4],[-.5,-2.5],[.5,-2.5],[.5,-4]])

fig,ax = plt.subplots()
p1 = Polygon(y,facecolor='k')
ax.add_patch(p)
ax.add_patch(p1)


plt.plot(x,f,'k')
plt.plot([-2,2], [-4,-4], 'k')
plt.axis('off')
plt.show()
```
