```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5
plt.plot(x,y,'kp')
plt.axis('square')
plt.axis([-6,6, -6,6])
plt.grid()
plt.show()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o', label="point %s" %(1+i))

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4,2,'r+')
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0],5.34343])
plt.show()
```

```python
import sympy as sym
```

```python
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-11,12):
    plt.plot(pnti, y.subs(x,pnti),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s'%y)
plt.show()
```
