```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
import scipy as sp
import matplotlib.pyplot as plt
```

```python
h = .999
t = np.linspace(-3,3,1001)

g = np.exp(-4*np.log(2)*t**2/h**2)

plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWHM ' + str(h))
plt.show()
```

```python
yval2find = .5
tvals = h*np.sqrt(np.log(yval2find)/(-4*np.log(2)))
print(tvals, -tvals)
```
