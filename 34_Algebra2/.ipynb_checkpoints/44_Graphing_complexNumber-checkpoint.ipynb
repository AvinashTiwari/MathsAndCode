```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
import matplotlib.pyplot as plt
```

```python
z = np.complex(2,3)
plt.plot(np.real(z), np.imag(z),'ro')
plt.plot([0,np.real(z)],[0, np.imag(z)], 'r')
plt.xlabel('real')
plt.ylabel('img')
plt.grid()
plt.axis([-4,4,-4,4])
```

```python
z1 = np.complex(-3,1)
z2 = np.complex(-1,1)

z1pz2 = z1+z2
plt.plot([0,np.real(z1)],[0, np.imag(z1)], label='z1')
plt.plot([0,np.real(z2)],[0, np.imag(z2)], label='z2')
plt.plot([0,np.real(z1pz2)],[0, np.imag(z1pz2)], label='z1pz2')
plt.xlabel('real')
plt.ylabel('img')
plt.legend()
plt.axis('square')
plt.grid()


plt.axis([-5,5,-5,5])
plt.show()
```

```python
z1tz2 = z1*z2
plt.plot([0,np.real(z1)],[0, np.imag(z1)], label='z1')
plt.plot([0,np.real(z2)],[0, np.imag(z2)], label='z2')
plt.plot([0,np.real(z1tz2)],[0, np.imag(z1tz2)], label='z1tz2')
plt.xlabel('real')
plt.ylabel('img')
plt.legend()
plt.axis('square')
plt.grid()


plt.axis([-5,5,-5,5])
plt.show()
```
