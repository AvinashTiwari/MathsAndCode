```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x =  np.linspace(-5,5,20)
feven = x**2
Fevennge = (-x)**2
plt.plot(x,feven)
plt.plot(x, Fevennge, 'ro')
plt.show()
```

```python
fOdd =x**3
fOddNeg  = (-x)**3
fNegOdd = -fOddNeg
plt.plot(x,fOdd)
plt.plot(x, fOddNeg, 'ro')
plt.plot(x, fOddNeg, 'g')

plt.show()
```

```python
theta = np.linspace(-2*np.pi, 2*np.pi,39)
cosfun = np.cos(theta)
cosfunN = np.cos(-theta)

plt.plot(theta,cosfun)
plt.plot(theta,cosfunN, 'ro')
plt.title('Cosin(x)')
plt.show()
```

```python
sinfun = np.sin(theta)
sinfunN = np.sin(-theta)
 
plt.plot(theta,sinfun)
plt.plot(theta,sinfunN, 'ro')
plt.title('Sin(x)')
plt.plot(theta,-sinfun)

plt.show()
```
