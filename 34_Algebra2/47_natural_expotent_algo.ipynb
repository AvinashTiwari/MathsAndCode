```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3, 99)
```

```python
plt.plot(x, np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.show()

plt.plot(x, np.log(x))
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
x = np.linspace(0.0001,10,10)
y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x, y1,label='log(e^x)')
plt.plot(x, y2, 'o',label='e^log(x)')
plt.legend()
plt.show()
```
