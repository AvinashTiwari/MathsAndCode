```python
x = np.linspace(-np.pi, 2*np.pi,40)
y = np.linspace(0, 4*np.pi,72)

X,Y =  np.meshgrid(x,y)

Fxy = np.cos(X) + np.sin(Y)
plt.imshow(Fxy)
plt.show()
```

```python
x = np.linspace(-2,2,100)
X,Y  = np.meshgrid(x,x)
s = 2
gause2d = np.exp(-(X**2+Y**2)/s)
plt.imshow(gause2d)
plt.axis('off')
plt.show()
```
