```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
import scipy as sp
import matplotlib.pyplot as plt
```

```python
a = 1
h = -10
k = -2
n = 100
x = np.linspace(h-5,h+5, n)

y = a*(x-h)**2 + k
```

```python
plt.plot(x,y)
plt.grid()
plt.axis('square')
plt.show()
```

```python
a = 1
p = 1/(4*a)
h = 1
k = -2
n = 100
x = np.linspace(h-2,h+2, n)

y = a*(x-h)**2 + k

plt.plot(x,y, label='parabola')
plt.plot(h,k, 'ro', label='Vertex')
plt.plot(h,k+p, 'go', label='Focus')
d = k-p
plt.plot(x[[0,-1]],[d,d], label='Diretix')
plt.grid()
plt.legend()
plt.show()
```
