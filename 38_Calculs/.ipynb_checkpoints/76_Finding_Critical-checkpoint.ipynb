```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
from scipy.signal import find_peaks
```

```python
x = np.linspace(-5,5,1000)
fx = x**2 * np.exp(-x**2)
dfx = np.diff(fx)/(x[1]-x[0])
localmax = find_peaks(fx)[0]
localmin =  find_peaks(-fx)[0]
print('Critical points are ' +str(x[localmax]) + "  " + str(x[localmin]))
plt.plot(x,fx)
plt.plot(x[0:-1],dfx)
plt.plot(x[localmax],fx[localmax],'ro' )
plt.plot(x[localmin],fx[localmin],'gs' )

plt.show()
```

```python
x = sym.symbols('x')
fx = x**2*sym.exp(-x**2)
dfx = sym.diff(fx)
criticalpoints = sym.solve(dfx)
criticalpoints
```

```python
a, x = sym.symbols('a,x')
baseexpr = x**2 * sym.exp(-a*x**2)
arange = np.arange(0,2.25,.25)
xrange = np.linspace(-3,3,100)
fig,ax = plt.subplots(1,2)

for ai in arange:
    fx = baseexpr.subs(a,ai)
    dfx = sym.diff(fx)
    critpnts = sym.solve(dfx)
    
    ax[0].plot(xrange,sym.lambdify(x,fx)(xrange))
    ax[1].plot(xrange,sym.lambdify(x,dfx)(xrange))
    
    if 1 in critpnts:
        display(Math("\\Rightarrow %s \\text{has a critical point x!!}" %sym.latex(fx)))
    elif 2 in critpnts:
        display(Math("\\Rightarrow %s \\text{has a critical point x!!}" %sym.latex(fx)))
    else:
        display(Math("\\quad %s \\text{has No critical point x!!}" %sym.latex(fx)))
        
        
    
ax[0].set_ylim([0,2])
ax[0].set_title('Function')
ax[0].plot([1,1],[0,2], '--', 'gray')
ax[0].plot([2,2],[0,2], '--', 'gray')


ax[1].set_ylim([-2,2])
ax[1].set_title('Its deravative')
ax[1].plot([1,1],[-2,2], '--', 'gray')
ax[1].plot([2,2],[-2,2], '--', 'gray')
```
