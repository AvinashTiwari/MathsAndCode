```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x =  sym.symbols('x')

f = x**2
df = sym.diff(f)

xa = 1

fa = f.subs(x,xa)
dfa = df.subs(x,xa)
dfa
```

```python
xx = np.linspace(-2,2,2000)
f_fun = sym.lambdify(x,f)(xx)
df_fun = sym.lambdify(x,df)(xx)

tanline = dfa *(xx-xa) + fa
plt.plot(xx,f_fun, label="f(x)")
plt.plot(xx, tanline, label="tangent")
plt.plot(xa,fa , 'ro')
plt.axis('square')
ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0], 'k--')
plt.plot([0,0],ax.get_ylim(), 'k--')
plt.legend()
plt.axis([-2,2,-.5,2])
```

```python
def computetangent(f,fa,bnds):
    df = sym.diff(f)
    fa = f.subs(x,xa)
    dfa = df.subs(x,xa)
    
    xx = np.linspace(bnds[0], bnds[1],200)
    return dfa *(xx-xa) + fa
    
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-2,2,200)
ffun = sym.lambdify(x,f)(xx)
tanline = computetangent(f,1,xx[[0,-1]])
plt.plot(xx,ffun)
plt.plot(xx,tanline)
plt.show()
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-2,2,200)
ffun = sym.lambdify(x,f)(xx)

for a in xx:
    y = computetangent(f,a, xx[[0,-1]])
    plt.plot(xx,y)

plt.plot(xx,ffun)
plt.show()
```
