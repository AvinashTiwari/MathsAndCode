```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
q = sym.symbols('q')
print(sym.diff(sym.cos(q)))
```

```python
print(sym.diff(sym.sin(q)))
```

```python
f = sym.cos(q)

for i in range(0,8):
    display(Math('\\frac{d}{dx}%s = %s ' %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot
f =  sym.cos(q)

for i in range(0,4):
    if i ==0:
        p = symplot(f, show=False, label=sym.latex(f), line_color=(i/5,i/5,i/5))
    else:
        p.extend(symplot(f, show=False, label=sym.latex(f),line_color=(i/5,i/5,i/5)))
    f = sym.diff(f)
    
p.legend = True
p.xlim = [-3,3]
p.show()
```

```python
x,a = sym.symbols('x,a')
f = sym.cos(x + sym.sin(x)) + a
colors = 'brkm'
for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai), show=False,label='a=' + str(ai), line_color = colors[0])
    else:
        p.extend(symplot(f.subs(a,ai), show=False,label='a=' + str(ai),line_color = colors[ai]))
        
p.title = "The functions"
p.legend = True
p.xlim = [-3,3]
p.show()
```

```python
for ai in range(0,4):
    if ai==0:
        p = symplot(sym.diff(f.subs(a,ai)), show=False,label='a=' + str(ai), line_color = colors[0])
    else:
        p.extend(symplot(sym.diff(f.subs(a,ai)), show=False,label='a=' + str(ai),line_color = colors[ai]))
        
p.title = "The deriverites"
p.legend = True

p.show()
```
