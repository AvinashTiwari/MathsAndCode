```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open("einstein.jpg")
```

```python
print(np.shape(ein))
```

```python
ein = np.mean(ein,2)
plt.imshow(ein)
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)
print(np.shape(U))
print(np.shape(s))
print(np.shape(V))
```

```python
fig,ax = plt.subplots(1,3)
ax[0].imshow(U)
ax[0].set_title('U')
ax[1].imshow(np.diag(s))
ax[1].set_title('$\\Sigma$')
ax[2].imshow(V)
ax[2].set_title('V')
```

```python
plt.plot(s, 'ks-', markerfacecolor='w')
plt.xlim([-1,50])
plt.title('EigenSpectrum on Einsten')
plt.show()
```

```python
S = np.zeros(np.shape(ein))

for i in range(0, len(s)):
    S[i,i] = s[i]

rein = U@S@V
plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('orginal')


plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('Svd version')

plt.show()
```

```python
renderorder = np.random.permutation(len(s))

S = np.zeros(np.shape(ein))

for i in range(0, len(s)):
    S[i,i] = s[renderorder[i]]

rein = U@S@V
plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('orginal')


plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('Svd version')

plt.show()
```
