```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open("einstein.jpg")
```

```python
print(np.shape(ein))
```

```python
ein = np.mean(ein,2)
plt.imshow(ein)
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)
print(np.shape(U))
print(np.shape(s))
print(np.shape(V))
```

```python
fig,ax = plt.subplots(1,3)
ax[0].imshow(U)
ax[0].set_title('U')
ax[1].imshow(np.diag(s))
ax[1].set_title('$\\Sigma$')
ax[2].imshow(V)
ax[2].set_title('V')
```
