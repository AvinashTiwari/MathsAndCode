```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open("einstein.jpg")
```

```python
print(np.shape(ein))
```

```python
ein = np.mean(ein,2)
plt.imshow(ein)
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)
print(np.shape(U))
print(np.shape(s))
print(np.shape(V))
```
